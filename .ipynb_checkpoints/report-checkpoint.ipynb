### Feature Engineering <a class="anchor" id="section_1_2"></a> 
#### Features used for clustering
- followers count
- friends count
- followers-friends ratio
- proportion of tweets
- proportion of RT
- proportion of replies
- days since account creation
- average number of statuses per month
- average number of hashtags per post
- average number of URLs per post
- average number of mentions per post
- average number of retweets per tweet

```python
def extract_info(df):
    
    n = len(df)
    n_replies = n - np.sum(df['replyto_userid'].isnull())
    n_rt = n - np.sum(df['retweeted_userid'].isnull())
    mentions = df[df['mentions_ids'].notnull()].mentions_ids
    n_mentions = len([x for xs in mentions for x in xs.split(',')])
    
    retweeted = np.nansum(df['retweet_count'])
    
    # number of hashtags
    hashtags = df[df['hashtags'].notnull()].hashtags
    n_hashtags = len([x for xs in hashtags for x in xs.split(',')])
    
    # number of urls per tweet
    urls = df[df['urls'].notnull()].urls
    n_urls = len([x for xs in urls for x in xs.split(',')])
    
    return n, n_replies, n_mentions, n_rt, n_hashtags, n_urls, retweeted
```

```python
# as the authors did in the paper, we will stick to Ego data points
isEgo = egoalter.loc[egoalter['IsEgo']]
```

```python
# we make sure the IDs used are both Egos and are of people that tweeted
ego_intersect = set(egotimes['egoID'].unique()).intersection(set(isEgo['ID']))
columns = ['id','n_statuses', 'n_replies','n_mentions','n_rt','n_hashtags','n_urls','retweeted']
stats = pd.DataFrame(0, index=np.arange(len(ego_intersect)), columns=columns)


for i, _id in tqdm(enumerate(ego_intersect)):
    n, n_replies, n_mentions, n_rt, n_hashtags, n_urls, retweeted = extract_info(egotimes.loc[egotimes['egoID']==_id])
    stats.iloc[i].id = _id
    stats.iloc[i].n_statuses = n
    stats.iloc[i].n_replies = n_replies
    stats.iloc[i].n_mentions = n_mentions
    stats.iloc[i].n_rt = n_rt
    stats.iloc[i].n_hashtags = n_hashtags
    stats.iloc[i].n_urls = n_urls
    stats.iloc[i].retweeted = retweeted
```

```python
merged = isEgo.merge(stats, how='left', left_on='ID', right_on='id')
merged.head()
```

```python
# computing features and merging them
inactive = (merged.friends_count == 0) & (merged.followers_count == 0)
merged['followers_ratio'] = (merged['followers_count']/(merged['friends_count'] + 0.01))
merged.loc[inactive,'followers_ratio'] = 1

merged['n_tweets'] = merged['n_statuses'] - merged['n_replies'] - merged['n_rt']
merged['prop_tweets'] = merged['n_tweets']/merged['n_statuses']
merged['prop_replies'] = merged['n_replies']/merged['n_statuses']
merged['prop_rt'] = merged['n_rt']/merged['n_statuses']
merged['avg_hashtags'] = merged['n_hashtags']/merged['n_statuses']
merged['avg_urls'] = merged['n_urls']/merged['n_statuses']
merged['avg_mentions'] = merged['n_mentions']/merged['n_statuses']
merged['avg_retweeted'] = (merged['retweeted']/merged['n_tweets']).replace(np.inf, 0)
merged['created_at'] = pd.to_datetime(merged['created_at'], format='%a %b %d %H:%M:%S %z %Y')
end_date = merged.created_at.max()
merged['account_age_months'] = ((end_date - merged.created_at).dt.days/30) + 0.01
merged['statuses_per_month'] = merged['statuses_count']/merged['account_age_months']
merged.head()
```

### DBSCAN <a class="anchor" id="section_2_1"></a> 

Since, our base assumption was that there are six main groups of the users, let's try clustering that will help us finding this 6 groups. 
First of all, we decided to try DBSCAN for clustering, because it does not have predifined amount of clusters and will be more honest detection for real amount of users groups.

```python
def find_average_user(X, labels):
    uniq_labels = sorted(list(set(labels)))
    average_users = []
    average_labels = []
    for label in uniq_labels:
        inds = [labels == label]
        inds = inds[0]
        cluster_users = X.iloc[inds]
        if len(cluster_users) > 100:
            average_user = cluster_users.mean(axis=0)
            average_users.append(average_user)
            average_labels.append(label)
    return dict(zip(average_labels, average_users))
```

```python
# We made a grid search to find the best parameters for DBSCAN, but it is too heavy to include it in the report
# We chose the best parameters based on silhouette score and number of clusters 
# (because cases where there is only 1 or 2 clusters does not work for us, since we will not conclude anything)
from sklearn.metrics import silhouette_score
eps = 1.
min_sample = 100
dbscan = DBSCAN(eps=eps, min_samples=min_sample).fit(norm_df)
unique_labels = set(dbscan.labels_)
sl = silhouette_score(norm_df, dbscan.labels_)
print(eps, min_sample, len(unique_labels), sl)
print(Counter(dbscan.labels_))
```

```python
unique_labels = set(dbscan.labels_) #get labels for DBSCAN
# -1 label means the cluster of points which was not include to any other cluster 
unique_colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
color_dict = dict(zip(unique_labels, unique_colors))
labels = dbscan.labels_
colors = []
for label in labels:
    colors.append(color_dict[label])
print(unique_labels)
```

### K-Means <a class="anchor" id="section_2_1"></a> 

To decide on the number of clusters to pick in a non-arbitrary way, we use the elbow method coupled with the distortion distance to the centers. 

```python
# source code from
# https://www.geeksforgeeks.org/elbow-method-for-optimal-value-of-k-in-kmeans/
distortions = [] 

X = norm_df.copy()
K = range(1,10) 
for k in K: 
    #Building and fitting the model 
    kmeanModel = KMeans(n_clusters=k,random_state=random_state).fit(X) 
    dist_cluster = sum(np.min(cdist(norm_df, kmeanModel.cluster_centers_, 
                      'euclidean'),axis=1)) / X.shape[0]
    distortions.append(dist_cluster) 
```

```python
plt.plot(K, distortions, 'bx-') 
plt.xlabel('Values of K') 
plt.ylabel('Distortion') 
plt.title('The Elbow Method using Distortion') 
plt.show() 
```

From the previous plot, we decide on picking $k=6$.

```python
k_opt = 6
kmeans = KMeans(n_clusters=k_opt,random_state=random_state).fit(norm_df)
centroids = kmeans.cluster_centers_
# centroids
```

```python
# find index of the centroids
def ID_centroids(centroids):
    # global defined features
    IDs = []
    for center in centroids:
        index = np.argmin(abs(np.sum((features.iloc[:,1:].to_numpy()-center),
                             axis=1)))
        IDs.append(features['ID'][index])
    
    return IDs
centres = ID_centroids(centroids)
```

```python
users = find_average_user(features.iloc[:, 1:], kmeans.labels_)
users_list = []
for k, v in users.items():
    users_list.append(v.values)
users_list = np.array(users_list)
users_list[:, 1]
```

### K-Means Visualization <a class="anchor" id="section_3_2"></a>   

```python
colors = ['red','black','blue','green','yellow', 'pink', 'magenta', 'cyan', 'grey', 'lightgreen']
LABEL_COLOR_MAP = {0:'r', 1:'k', 2:'b', 3:'g', 4:'y', 5:'pink', 6:'m',7:'c', 8:'grey', 9:'lightgreen'}
label_color = [LABEL_COLOR_MAP[l] for l in kmeans.labels_]

#orthogonal projection, onto space spanned by two
def plot_project1d(I,features, k):
    #idx(i) globally defined
    fig, axs = plt.subplots(4,3,figsize=(15,15))
    
    for index, i in enumerate(I):
        m,s = index%4, index//4
        for c in range(k):
            try:
                features.iloc[idx[c],i].plot.kde(label=c, ax=axs[m,s], c=colors[c])
            except:
                axs[m,s].axvline(features.iloc[idx[c],i].median(), 0,1, label=c, c=colors[c])
        axs[m,s].legend()
        axs[m,s].set_yscale('symlog')
        axs[m,s].set_xscale('symlog')
        axs[m,s].set_ylim([0, 1])
        axs[m,s].title.set_text('{}'.format(features.columns[index]))
    fig.tight_layout()
    
def plot_project2d(pairs,features,centroids):
    fig, axs = plt.subplots(4,3,figsize=(15,15))
    count = 0
    for i,j in zip(pairs[0],pairs[1]):
        m,s = count%4, count//4
        for g in np.unique(kmeans.labels_):
            axs[m,s].scatter(centroids[g, i], centroids[g, j], c=colors[g], label=g, s=100)
        axs[m,s].legend()
        axs[m,s].title.set_text('{} vs {}'.format(features.columns[i],features.columns[j]))
        count += 1
        
    fig.tight_layout()
```

## Evolution of profiles <a class="anchor" id="chapter4"></a>  

```python
def distance(a, b):
        return np.sqrt(np.sum((a - b)**2, axis=0))

def get_distance_from_centre(df, centroids):
    distances = []
    for i in range(0, len(df)):
        d = distance(centroids[kmeans.labels_[i]], df.iloc[i,:].values)
        distances.append(d)
    return distances

df_dist = norm_df.copy()
df_dist['dist'] = get_distance_from_centre(norm_df, centroids)
df_dist['cluster'] = kmeans.labels_

# get the closest 5 points from the cluster centre
close = df_dist.groupby('cluster', group_keys=False).apply(
    lambda x: x.sort_values('dist', ascending=True).head(10))
close.index.name = 'ID'
close_tocenters = close.reset_index()
```

```python
# we do a daily activity here -> can change feature to be more than activity
def find_id_evolution(ID,verbose=False):
    # global egotimes
    id_evolution = egotimes.loc[egotimes['egoID']==ID]
    id_evolution['created_at'] = pd.to_datetime(id_evolution['created_at'], 
                                            format='%a %b %d %H:%M:%S %z %Y')

    if not len(id_evolution['created_at']):
        if verbose: print('Only one day of activity or none at all')
        return
    if (max(id_evolution['created_at']) - min(id_evolution['created_at'])).days <=1:
        if verbose: print('Only one day of activity or none at all')
        return 
    
    id_evolution['n-th_day']=(id_evolution['created_at']-min(id_evolution['created_at'])).apply(lambda x: x.days)
    return id_evolution

# might be useful in some interactive plots
def plot_id_evolution(ID):
    id_evolution = find_id_evolution(ID)
    if id_evolution is None:
        ax = plt.plot()
    else:
        ax = id_evolution.groupby('n-th_day').nunique()['tweetid'].plot(label='number of tweets')
        ax = id_evolution.groupby('n-th_day').nunique()['urls'].plot(label='url usage')
        ax = id_evolution.groupby('n-th_day').nunique()['retweet_count'].plot(label='retweet count')
#    ax.set_ylabel('Number of tweets')
        ax.set_xlabel('n-th day since recorded first tweet')
        ax.set_yscale('log')
        ax.legend()


def plot_centroid_evolution(id_centroids,distances):
    # k_opt again global
    n = len(distances)/k_opt + 1 # number of points that we average over
    fig, axs = plt.subplots(3,2,figsize=(20,25))

    for index,id_center in enumerate(id_centroids):
        m,s = index%3, index//3
        
        neighbour_id = list(distances.loc[distances['cluster']==index]['ID'])
        
        Tweet_evolution = {}
        Urls_evolution = {}
        Retweet_evolution = {}
        for ID_pts in (neighbour_id + [id_center]):
            id_df = find_id_evolution(ID_pts)
            if id_df is None: 
                continue
            tweet_evolution = id_df.groupby('n-th_day').nunique()['tweetid'].to_dict()
            urls_evolution = id_df.groupby('n-th_day').nunique()['urls'].to_dict()
            retweet_evolution = id_df.groupby('n-th_day').nunique()['retweet_count'].to_dict()
            
            for key,val in tweet_evolution.items():
                if key in Tweet_evolution:
                    Tweet_evolution[key] += val
                    Urls_evolution[key] += urls_evolution[key]
                    Retweet_evolution[key] += retweet_evolution[key]
                else:
                    Tweet_evolution[key] = val
                    Urls_evolution[key] = urls_evolution[key]
                    Retweet_evolution[key] = retweet_evolution[key]

        if len(Tweet_evolution) <= 2:
            axs[m,s] = plt.plot()
#            axs[m,s][0].title.set_text('centroid #{}'.format(index))
        else:
            
            Tweet_evolution = dict(sorted(Tweet_evolution.items()))
            Urls_evolution = dict(sorted(Urls_evolution.items()))
            Retweet_evolution = dict(sorted(Retweet_evolution.items()))
            
            counts = {'n-th_day':list(Tweet_evolution.keys()),
                     'number_of_tweets':np.array(list(Tweet_evolution.values()))/n,
                     'url_usage':np.array(list(Urls_evolution.values()))/n,
                     'retweet_count':np.array(list(Retweet_evolution.values()))/n}
            
            fulldf = pd.DataFrame(data=counts)
            fulldf = fulldf.set_index('n-th_day')
            axs[m,s] = fulldf.plot(ax=axs[m,s])
            
            axs[m,s].set_ylim([0, 10])
            axs[m,s].set_yscale('symlog')
            axs[m,s].title.set_text('centroid #{}'.format(index))
```

```python
plot_centroid_evolution(centres, close_tocenters)
```

The plots shows the average twitter activity of the closest 10 points to the centroids of clusters. 
0. The average users seem to be active only first three months and then lose interest. 
1. The active users for certain period.
2. Comapatively low activity and higher usage of url towards the end of period. 
3. Linear declining trend in the activity.
4. Rare visitors, probably tweets on holidays.
5. Regular active users.
